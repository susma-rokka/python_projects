##### Data Visualization

```python
# Histogram
plt.figure(figsize=(8,6))
sns.histplot(df['Unit price'], kde=True)
plt.title('Distribution of Tax')
plt.xlabel('Unit Price')
plt.ylabel('Tax 5%')
plt.show()
```

```python
# Histogram of customer ratings
plt.figure(figsize=(10, 6))
sns.histplot(df['Rating'], bins=10, kde=True)
plt.title('Customer Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# Group by product line and calculate total sales
sales_by_product_line = df.groupby('Product line')['Total'].sum().reset_index()

# Bar plot for sales by product line
plt.figure(figsize=(10, 6))
sns.barplot(x='Total', y='Product line', data=sales_by_product_line)
plt.title('Sales by Product Line')
plt.xlabel('Total Sales')
plt.ylabel('Product Line')
plt.show()
```

```python
# Bar Graph
plt.figure(figsize=(8, 6))
sns.countplot(x='Gender', data=df)
plt.title('Count of Customers by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
df.columns
```

```python
df['Date'] = pd.to_datetime(df['Date'])

# Group by date and calculate total sales
sales_over_time = df.groupby('Date')['Total'].sum().reset_index()

# Line plot for sales over time
plt.figure(figsize=(10, 6))
sns.lineplot(x='Date', y='Total', data=sales_over_time)  # Use 'data' instead of 'df'
plt.title('Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(8, 6))
sns.boxplot(x='Product line', y='Unit price', data=df)
plt.title('Unit Price by Product line')
plt.xlabel('Product line')
plt.ylabel('Unit price')
plt.xticks(rotation=45)
plt.show()
```

```python

gender_counts = df['Gender'].value_counts()
# print(gender_counts)

# Create a pie chart
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightpink'])
plt.title('Distribution of Gender')
plt.show()
```
